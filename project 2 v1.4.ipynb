+ Defining the function kMeans to perform the task of clustering.  

The method followed is of Eucledian distance. The distance is then calculated of each centroid from Frequency, Amount and Recency. Based on this distance, clusters are formed. The mean of newly formed clusters become the centroids for the next iteration.

```python
def kMeans(final_norm_km,centroid1,centroid2,centroid3):
    #claculating the eucaldian distance for three centroids with the datapoints
    final_norm_km["Dist_C1"] = np.sqrt((final_norm_km.Frequency - centroid1[0])**2 + (final_norm_km.Amount - centroid1[1])**2 + (final_norm_km.Recency - centroid1[2])**2)
    final_norm_km["Dist_C2"] = np.sqrt((final_norm_km.Frequency - centroid2[0])**2 + (final_norm_km.Amount - centroid2[1])**2 + (final_norm_km.Recency - centroid2[2])**2)
    final_norm_km["Dist_C3"] = np.sqrt((final_norm_km.Frequency - centroid3[0])**2 + (final_norm_km.Amount - centroid3[1])**2 + (final_norm_km.Recency - centroid3[2])**2)
 
    #Clustering the datapoints according to the distance 
    final_norm_km["Cluster"] = np.where((final_norm_km.Dist_C1 < final_norm_km.Dist_C2) & 
                                  (final_norm_km.Dist_C1 < final_norm_km.Dist_C3), 1, 
                                   np.where((final_norm_km.Dist_C2 < final_norm_km.Dist_C1) & 
                                  (final_norm_km.Dist_C2 < final_norm_km.Dist_C3) , 2, 3))
    
    #caluclating the mean of new centroids with the associated datapoints
    centroid1 = final_norm_km[final_norm_km.Cluster == 1][["Frequency", "Amount", "Recency"]].mean()
    centroid2 = final_norm_km[final_norm_km.Cluster == 2][["Frequency", "Amount", "Recency"]].mean()
    centroid3 = final_norm_km[final_norm_km.Cluster == 3][["Frequency", "Amount", "Recency"]].mean()
    
    return (final_norm_km,centroid1,centroid2,centroid3)
```

A loop is created to run iterations until the centroids dont move further, in other words until value of centroids dont change.The loop returns number of iterations performed and the final centroids.

```python
n = range(50)

for i in n:
    final_norm_km,p,q,r= kMeans(final_norm_km,centroid1,centroid2,centroid3)
    if(np.array_equal(centroid1,p)& np.array_equal(centroid2,q) & np.array_equal(centroid3,r)):
        break
    else: 
        centroid1 = p
        centroid2 = q
        centroid3 = r
        #print(i)
print('Iterations :', i)
print ('Final Centroids are ') 
print(centroid1,centroid2,centroid3)
```

The method followed here is clustering and the output is a class membershop or cluster memebrship.  A datapoint is clustered by a plurality vote of its neighbors. The datapoint is assigned to the cluster that is mos common among its k nearest neighbors. For the project K-NN  is perfomed on Amount, Recency and Frequency. There are 3 clusters. When a new datapoint is added to the cluster, an insight will be provided on where exactly the new datapoint or the customer fits in and can be used in the existing target campaign /advertisement.
